## 2.1 Read or convert the columns ‘Created Date’ and Closed Date’ to datetime datatype

```python
def date_time_conversion(string): # function to convert various string formats to datetime
    try:
        return datetime.strptime(str(string),"%m/%d/%Y %H:%M:%S %p")


    except (ValueError,TypeError) as e:
        try:
            return datetime.strptime(string,"%m/%d/%y %H:%M")
        except (ValueError,TypeError) as e:
            return datetime.strptime(str(string),"%m-%d-%y %H:%M")
```

```python
# Reformatting 'Date Created' Column and "Closed Date"
df["Created Date"]=df["Created Date"].apply(date_time_conversion)
```

```python
df["Closed Date"]=df["Closed Date"].apply(date_time_conversion)
```

## 2.2 Create a new column ‘Request_Closing_Time

```python
 df['Request_Closing_Time']=df["Closed Date"]-df["Created Date"]
```
